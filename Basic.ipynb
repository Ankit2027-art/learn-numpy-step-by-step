📌Difference Between LIST
 and ARRAY

| Feature                 | List                      | NumPy Array         |
| ----------------------- | ------------------------- | ------------------- |
| Library Needed          | No                        | Yes (`NumPy`)       |
| Speed                   | Slower                    | Faster              |
| Memory Usage            | More memory               | Less memory         |
| Data Type               | Can store different types | Usually same type   |
| Mathematical Operations | Difficult                 | Easy                |
| Performance             | Good for small data       | Best for large data |
